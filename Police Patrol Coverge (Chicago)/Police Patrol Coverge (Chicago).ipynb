# Where Should Police Patrols be Located in Chicago To Minimize Crime?

# Introduction and Thesis Statement

Chicago is the third-largest city in the United States, yet it is well known for being one of the
most dangerous in the country. Its violent crime rate is 164% higher than the national average
and its property crime rate is 45% higher. My project is on reducing crime by
optimally stationing police patrols on the most common areas that crime occurs.

# Data

The data is from https://www.chicago.gov/city/en/dataset/crime.html which they extracted from
the Chicago Police Department's CLEAR (Citizen Law Enforcement Analysis and Reporting)
system. The data set is on all crimes reported to the Chicago Police Department. The data going
to be used from the data set is time, description of the crime, and location (in the form of latitude
and longitude). The data set obtained from the website spans from 00:00 12/08/2020 to 00:00 12/08/2021.
Since it is assumed that crimes for different seasons and months are different, I cleaned the data to the time spans from 00:00 11/08/2020 to 00:00 12/08/2021 (1 month). User can input data from any time frame into the program to obtain a different optimal value.

# Approach

The optimization to be solved is a set covering problem. The problem would be focused on
maximizing crimes covered based on location. The crime to be covered would be grouped
based on latitude and longitude round to the 3 decimal points, for the data I used, that is 17436 crimes. The problem solved would be the optimal locations which would be chosen by latitude and longitude round to 2 decimal points, which I ended with 642 possible locations. A distance perimeter is set up, calculating the 1-norm distance between points (crime vs location of a patrol). The distance would be in kilometers. The default distance of a covered crime is within 1.5 km one norm of the patrol location. The user would be prompted to change this number if they wanted to change the parameter. The number of patrols would be based on a theoretical cost of having a patrol and a theoretical budget, both could be input by the user. The default inputted is 672900, which is the average salary of 10 police officers in a patrol. Type of crime and time of day can be filtered by user. 

Parameters:
1. Users can input cost of a patrol
2. Users can input the budget of police deparment for patrols
3. Users can input the distance of a crime from a location to be considered covered.
4. Users can filter by type of crime
5. Users can filter by time of day

Outputs:
1. Optimal Value: Number of and percentage of crimes covered
2. Plot of optimal locations to send patrols on a Chicago map
3. Plot of crimes covered vs uncovered on a Chicago map

Creating a matrix with possible patrol locations on the y axis and crimes on the x axis. Calculating the distance each of the patrol location and crime locations. The distance obtained was in kilometer, using the formula (Which the numbers are only accurate for Chicago: |Lat_Patrol-Lat_Crime| * 111.2 +|Lon_Patrol-Lon_Crime| * 82.63

```python
Matrix_list=[]
for i in range(len(Location_df)):
    Matrix_list_sub=[]
    for j in range(len(Crime_df)):
        Matrix_list_sub.append(abs(Location_df[0][i]-Crime_df[0][j])*111.2+abs(Location_df[1][i]-Crime_df[1][j])*82.63)
    Matrix_list.append(Matrix_list_sub)
    
```

```python
Matrix_df=pd.DataFrame(Matrix_list)
```

Creating a binary matrix that determined if a crime is covered by each of the patrol locations. Value is 1 if distance is within D (Value inputed by user), 0 if not.

```python
for i in range(17436):
    for j in range(642):
        if Matrix_df[i][j]-D<0:
            Matrix_df[i][j]=1
        else:
            Matrix_df[i][j]=0
Matrix_df
```

```python
%%gams
$title Final_Project
option limrow=0, limcol=0;
option solprint=off;
*$call csv2gdx Location.csv id=location index=1 values=2..lastCol useHeader=y
*call csv2gdx Crime.csv id=crime index=1 values=2..lastCol useHeader=y
$call csv2gdx Matrix.csv id=covered index=1 values=2..lastCol useHeader=y
*$call csv2gdx Matrix.csv id=covered index=1..2 values=lastCol useHeader=y
$call csv2gdx parameters_final.csv id=inputs index=1 values=2..lastCol useHeader=y

sets
i
j
xp
yp
;

parameters
Percentage_covered
covered(i,j)
inputs(xp,yp)
;

Scalar N,Budget,Cost_Patrol;


$gdxin parameters_final.gdx
$load xp=Dim1
$load yp=Dim2
$load inputs

$gdxin Matrix.gdx
$load i=Dim1
$load j=Dim2
$load covered

Budget=inputs("0","Budget");
Cost_Patrol=inputs("0","Cost_Patrol");
N=round(Budget/Cost_Patrol);

free variables Crime_covered;
binary variables
x(i) if there is patrol in the location
y(j) if those crime are covered

equations
obj
num_locations
Covering
;

obj..
Crime_covered =e= sum(j,y(j));

num_locations..
sum(i,x(i)) =e= N;

Covering(j)..
sum(i,covered(i,j)*x(i)) =g=  y(j);


model Final_Project /all/;
solve Final_Project using mip max Crime_covered;





```
